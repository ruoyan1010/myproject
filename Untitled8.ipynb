```python
from pyspark.sql import SparkSession
import os

# === 1. JDBC jar 路径（相对路径即可） ===
jar_path = "jars/mysql-connector-j-8.0.33.jar"
assert os.path.exists(jar_path), f"JAR 文件不存在：{jar_path}"

# === 2. 初始化 SparkSession ===
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("mysql-test") \
    .config("spark.jars", jar_path) \
    .getOrCreate()

print("Spark 版本：", spark.version)

# === 3. MySQL 连接参数（请根据你的 MySQL 设置修改 user/password） ===
jdbc_url = "jdbc:mysql://127.0.0.1:3306/testdb"

properties = {
    "user": "root",          # 如果你用的是 root，就写 root
    "password": "你的密码",   # 改成你自己的 MySQL 密码
    "driver": "com.mysql.cj.jdbc.Driver"
}

# === 4. 从 MySQL 读取 users 表 ===
df = spark.read.jdbc(
    url=jdbc_url,
    table="users",
    properties=properties
)

df.show()
print("总行数：", df.count())

spark.stop()
```

```python
from pyspark.sql import SparkSession
import traceback

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("debug-mysql") \
    .config("spark.jars", "jars/mysql-connector-j-8.0.33.jar") \
    .getOrCreate()

jdbc_url = "jdbc:mysql://127.0.0.1:3306/testdb?useSSL=false&serverTimezone=UTC"

properties = {
    "user": "root",
    "password": "123456",
    "driver": "com.mysql.cj.jdbc.Driver"
}

try:
    df = spark.read.jdbc(url=jdbc_url, table="users", properties=properties)
    df.show()
except Exception as e:
    print("="*80)
    print("PYTHON 侧错误：")
    traceback.print_exc()
    print("="*80)

    print("JAVA 侧错误：")
    try:
        je = e.java_exception
        print("Message:", je.getMessage())
        print("\n--- Java StackTrace (前50行) ---")
        for i, line in enumerate(je.getStackTrace()[:50]):
            print(i, line)
    except:
        print("无法获取 java_exception")

finally:
    spark.stop()
```
