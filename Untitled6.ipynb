```python
from pyspark.sql import SparkSession
import pandas as pd

# 1) 建 Spark（如果 notebook 里已有 spark session 可以跳过）
spark = SparkSession.builder.master("local[*]").appName("pandas-test").getOrCreate()

# 2) 创建一个小的 pandas DataFrame
pdf = pd.DataFrame({"id":[1,2,3,4,5], "val":["a","b","c","d","e"]})
print("pandas df:", pdf.shape)

# 3) 转成 Spark DataFrame
sdf = spark.createDataFrame(pdf)
sdf.show()

# 4) 再转回 pandas（注意：只做小数据）
pdf2 = sdf.toPandas()
print("back to pandas:", pdf2.shape)
print(pdf2.head())

# 5) 结束（可选）
# spark.stop()
```
