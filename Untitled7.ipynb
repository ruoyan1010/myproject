```python
from pyspark.sql import SparkSession
import os

# 相对路径或绝对路径都可以。假设你把 jar 放在项目根下的 jars/
jar_path = "jars/mysql-connector-j-8.0.33.jar"
assert os.path.exists(jar_path), f"找不到 jar: {jar_path}"

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("jdbc-test") \
    .config("spark.jars", jar_path) \
    .getOrCreate()

print("Spark version:", spark.version)
```
